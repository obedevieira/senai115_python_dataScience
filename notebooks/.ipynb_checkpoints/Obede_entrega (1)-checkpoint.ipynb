# Nova seção

```python
import pandas as pd

# Carregando os arquivos CSV em DataFrames separados
df_a = pd.read_csv('/content/Video_Game_Sales_a - Video_Game_Sales_a.csv')
df_b = pd.read_csv('/content/Video_Game_Sales_b - Video_Game_Sales_b.csv')
```

```python
# Verificação das colunas de dos DataFrames
print("Colunas em df_a:", df_a.columns)
print("Colunas em df_b:", df_b.columns)

# Verificação das primeiras linhas dos DataFrames
print(df_a.head())
print(df_b.head())
```

```python
# Combinando os DataFrames em um único DataFrame
df_combined = pd.concat([df_a, df_b], ignore_index=True)
```

```python
# Verificando o número de colunas no DataFrame
print(df_combined.columns)
print(len(df_combined.columns))
```

```python
# Renomeando as colunas para português
df_combined.columns = [
    'Classificação', 'Nome', 'Plataforma', 'Ano_Lancamento', 'Genero', 'Publicador',
    'Vendas_NA', 'Vendas_EU', 'Vendas_JP', 'Vendas_Outros', 'Vendas_Globais'
]
```

```python
#verificando as colunas
print(df_combined.columns)
display(df_combined.columns)
```

```python
display(df_combined.columns)
```

```python
# Exibindo as primeiras 5 linhas do DataFrame que foi combinado
print(df_combined.head())
display(df_combined.head())
```

```python
# Exibindo informações gerais do DataFrame
print(df_combined.info())
display(df_combined.info())

# Exibindo estatísticas descritivas do DataFrame
print(df_combined.describe())
display(df_combined.describe())
```

# 'Análise Exploratória'

```python
# quantidade de jogos por plataforma
quantidade_jogos_plataforma = df_combined['Plataforma'].value_counts()
print(quantidade_jogos_plataforma)
display(quantidade_jogos_plataforma)
```

```python
df_combined.columns = [
    'Classificação', 'Nome', 'Plataforma', 'Ano_Lancamento', 'Genero', 'Publicador',
    'Vendas_NA', 'Vendas_EU', 'Vendas_JP', 'Vendas_Outros', 'Vendas_Globais'
]

print(df_combined.columns)
```

```python
# Aqui tive que converter a coluna 'Vendas_Globais' para numérico, convertendo erros para NaN
df_combined['Vendas_Globais'] = pd.to_numeric(df_combined['Vendas_Globais'], errors='coerce')
```

```python
media_vendas_globais_plataforma = df_combined.groupby('Plataforma')['Vendas_Globais'].mean()
print(media_vendas_globais_plataforma)

display(media_vendas_globais_plataforma)
```

```python
# quantidade de jogos por gênero
quantidade_jogos_genero = df_combined['Genero'].value_counts()
print(quantidade_jogos_genero)
```

# Filtragem de dados

```python
# média das vendas globais
media_vendas_globais = df_combined['Vendas_Globais'].mean()

# jogos com vendas globais acima da média
jogos_acima_media = df_combined[df_combined['Vendas_Globais'] > media_vendas_globais]

# jogos acima da média
print(len(jogos_acima_media))
```

```python
# jogos lançados após o ano 2010
jogos_pos_2010 = df_combined[df_combined['Ano_Lancamento'] > 2010]

# média das vendas globais desses jogos
media_vendas_jogos_pos_2010 = jogos_pos_2010['Vendas_Globais'].mean()
print(media_vendas_jogos_pos_2010)
```

# Operações avançadas

```python
# Adicionando uma coluna ao DataFrame que calcule a soma das vendas na América do Norte e Europa
df_combined['Vendas_NA_EU'] = df_combined['Vendas_NA'] + df_combined['Vendas_EU']
```

```python
# Agrupando os dados por ano de lançamento e calcular a média das vendas globais
media_vendas_por_ano = df_combined.groupby('Ano_Lancamento')['Vendas_Globais'].mean()
print(media_vendas_por_ano)
display(media_vendas_por_ano)
```
