### Exercício: Explorando o Conjunto de Dados de Vendas de Videogames

**Objetivo:** Utilizar a biblioteca Pandas para extrair e analisar informações dos conjuntos de dados de vendas de jogos e video games.


### Tarefas dos Alunos

1. **Leitura e Inspeção dos Dados:**
   - Carregue os arquivos CSV em DataFrames separados.
   - Combine os DataFrames em um único DataFrame.
   - Renomeie as colunas para português
   - Exiba as primeiras 5 linhas do DataFrame combinado.
   - Exiba informações gerais e estatísticas descritivas do DataFrame.

2. **Análise Exploratória:**
   - Conte a quantidade de jogos por plataforma.
   - Calcule a média das vendas globais por plataforma.
   - Conte a quantidade de jogos por gênero.

3. **Filtragem de Dados:**
   - Filtre os jogos com vendas globais acima da média e conte quantos são.
   - Filtre os jogos lançados após o ano 2010 e calcule a média das vendas globais desses jogos.

4. **Operações Avançadas:**
   - Adicione uma coluna ao DataFrame que calcule a soma das vendas na América do Norte e Europa.
   - Agrupe os dados por ano de lançamento e calcule a média das vendas globais.



```python
#1.Leitura e Inspeção CSV em DataFrames separados.

##Carregue os arquivos CSV em DataFrames separados
import pandas as pd #importar a biblioteca para utilizar os métodos contidos nela

df_vendas_vg_a = pd.read_csv('/content/Video_Game_Sales_a.csv')
df_vendas_vg_b = pd.read_csv('/content/Video_Game_Sales_b.csv')

##Combine os DataFrames em um único DataFrame
df_vendas_vg = pd.concat([df_vendas_vg_a, df_vendas_vg_b])
#df_vendas_vg = df_vendas_vg_a + df_vendas_vg_b #(outra opção para analisar posteriormente)
print(df_vendas_vg.shape[0]) #qtd de linhas dos df concatenados
print(df_vendas_vg_a.shape[0]) #qtd de linhas do df_a
print(df_vendas_vg_b.shape[0]) #qtd de linhas do df_b
```

```python
##Renomeie as colunas para português
df_vendas_vg.columns = ['Rank', 'Nome', 'Plataforma', 'Ano', 'Genero', 'Empresa', 'AmericaNorte_Vendas',
                      'Europa_Vendas', 'Japao_Sales', 'Outras_Vendas', 'Vendas_Globais']

df_vendas_vg.columns
```

```python
##Exiba as primeiras 5 linhas do DataFrame combinado.
df_vendas_vg.head()
```

```python
##Exiba informações gerais e estatísticas descritivas do DataFrame.
print('=======================================================================')
print('informações tabela')
print(df_vendas_vg.info())
print('=======================================================================')
print('descrição tabela')
print(df_vendas_vg.describe())
```

```python
#2.Análise Exploratória:

##Conte a quantidade de jogos por plataforma.
df_vendas_vg_no_dup = df_vendas_vg.drop_duplicates() #deleta linhas completemante semelhantes
jogos_por_platforma = df_vendas_vg['Plataforma'].value_counts()
jogos_por_platform_no_dup = df_vendas_vg_no_dup['Plataforma'].value_counts()
print(jogos_por_platforma)
print(jogos_por_platform_no_dup)
```

```python

numero = '10;' #declarado como string

numero = numero.replace(';','')
print(numero)
#print(type(int(numero))) # string foi transformado em um inteiro
print(type(float(numero)))
```

```python
##Calcule a média das vendas globais por plataforma.
###tratar a coluna Vendas Globais
#df_vendas_vg['Vendas_Globais'] = df_vendas_vg['Vendas_Globais'].str.replace(';','')
#type(df_vendas_vg['Vendas_Globais'].str)
df_vendas_vg['Vendas_Globais'] = pd.to_numeric(df_vendas_vg['Vendas_Globais'])
df_vendas_vg.dtypes
medias_vendas_globais = df_vendas_vg.groupby('Plataforma')['Vendas_Globais'].mean()
medias_vendas_globais
```

```python
##Conte a quantidade de jogos por gênero.
jogo_por_genero = df_vendas_vg['Genero'].value_counts()
jogo_por_genero
```

```python
#3. Filtragem de Dados:
media_vendas_globais = df_vendas_vg['Vendas_Globais'].mean()
media_vendas_globais
#Filtre os jogos com vendas globais acima da média e conte quantos são.
jogos_venda_acima_media = df_vendas_vg.loc[(df_vendas_vg['Vendas_Globais'] > media_vendas_globais)]
#jogos_venda_acima_media.shape[0]
print(len(jogos_venda_acima_media))

#Filtre os jogos lançados após o ano 2010 e calcule a média das vendas globais desses jogos.
ano_filtro = 2010
jogos_venda_maior_2010 = df_vendas_vg.loc[(df_vendas_vg['Ano'] > ano_filtro)]
jogos_venda_maior_2010['Vendas_Globais'].mean()
```

```python
#4. Operações Avançadas:

#Adicione uma coluna ao DataFrame que calcule a soma das vendas na América do Norte e Europa.
df_vendas_vg['AN_EU_Vendas'] = df_vendas_vg['Europa_Vendas'] + df_vendas_vg['AmericaNorte_Vendas']
df_vendas_vg.columns
df_vendas_vg.head()

#Agrupe os dados por ano de lançamento e calcule a média das vendas globais.
media_vendas_globais_ano = df_vendas_vg.groupby('Ano')['Vendas_Globais'].mean()
media_vendas_globais_ano
```
