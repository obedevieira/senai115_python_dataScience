```python
#importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#ler o arquivo csv
df = pd.read_csv('/home/usuario/Área de trabalho/tattttttt/obede.csv')
```

```python
print(df.info())
```

```python
df_cleaned = df.dropna()

#apresenta 
df_cleaned.head()

print(df_cleaned)


print(df_cleaned.shape)
```

```python
mode_counts = df_cleaned['mode'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(mode_counts, labels=mode_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribuição dos Tipos de Veículos (mode)')
plt.show()
```

```python
region_counts = df_cleaned.groupby('region')['value'].sum().nlargest(10)

plt.figure(figsize=(10, 6))
region_counts.plot(kind='bar', color='skyblue')
plt.title('Top 10 Regiões em Quantidade de Veículos')
plt.xlabel('Região')
plt.ylabel('Quantidade de Veículos')
plt.xticks(rotation=45)
plt.show()
```

```python
df = pd.read_csv('/home/usuario/Área de trabalho/tattttttt/obede.csv')# Exibe as primeiras linhas do DataFrame para confirmar a importação
print(df.head())
```

```python
df.dropna()  # Remove todas as linhas com qualquer `NaN`
df.dropna(axis=1)  # Remove colunas com qualquer `NaN`
```

```python
df.isna().sum() 
```

```python
df.rename(columns={
    'region': 'regiao',
    'country': 'pais',
    'year': 'ano',
    'mode': 'tipo_veiculo',
    'value': 'quantidade'
}, inplace=True)

```

```python
print(df.info())
```

```python
df_cleaned = df.dropna()
```

```python
tipo_veiculo_counts = df_cleaned['tipo_veiculo'].value_counts()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(tipo_veiculo_counts, labels=tipo_veiculo_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribuição dos Tipos de Veículos')
plt.show()

```

```python
regiao_counts = df_cleaned.groupby('regiao')['quantidade'].sum().nlargest(10)
```

```python
plt.figure(figsize=(10, 6))
regiao_counts.plot(kind='bar', color='skyblue')
plt.title('Top 10 Regiões em Quantidade de Veículos')
plt.xlabel('Região')
plt.ylabel('Quantidade de Veículos')
plt.xticks(rotation=45)
plt.show()
```
